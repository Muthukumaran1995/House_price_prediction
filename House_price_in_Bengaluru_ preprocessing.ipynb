```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
def convert_sq_ft_into_number(x):
    token = x.split('-')
    if len(token) ==2:
        return (float(token[0]) + float(token[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
data['total_sqft'] = data['total_sqft'].apply(convert_sq_ft_into_number)
```
